```python
def load_data(file_name):
    with open(file_name,"r") as f:
        data=json.load(f)

    return data

 
```

```python
def clean_data(data):
    ch_to_num={"eighty five":85,"ninety":90,"seventy":70,"fifty five":55,"ninety two":92}
    unique_data=set()
    cleaned_data=[]

    for user in data:
        raw_marks=str(user["marks"]).strip().lower()
        if raw_marks in ch_to_num:
            raw_marks=ch_to_num[raw_marks]
            user["marks"]=raw_marks

        if user["name"].strip() in unique_data:
            continue

        unique_data.add(user["name"])
        cleaned_data.append(user)

        raw_age=user.get("age")
        if raw_age == None:
            user["age"]=None
        
    return cleaned_data
        
```

```python
def get_insights(data):
    insights=[]
    total_students=0
    for user in data:
        total_students += 1
    print(f"the total students are= {total_students}")

    total_marks=0
    marks=[]
    for user in data:
        marks.append(user["marks"])
        total_marks += user["marks"]
        highest_marks=max(marks)
        lowest_marks=min(marks)
        
        

    print(f"the avg marks= {total_marks / len(data)}")
    print(f"the highest marks is={highest_marks}")
    print(f"the lowest marks is={lowest_marks}")

    stu_marks_fifty=0
    for user in data:
        if user["marks"] < 50:
            stu_marks_fifty += 1

    print(f"{stu_marks_fifty} students who's marks less than 50")

    attend_less_75=0
    for user in data:
        if user["attendance"] < 75:
            attend_less_75 += 1

    print(f"{attend_less_75} students who's attendance less than 75")
        

        
    
```

```python
def get_recommendation(data):
    recommendation=[]
    for user in data:
        result={}
        result["name"]=user["name"]

        if user["marks"] >= 90:
            result["status"]="Excelent"

        elif user["marks"] >= 80:
            result["status"]="Best"

        elif user ["marks"] >= 70:
            result["status"] = "good"

        else:
            result["status"]="need improvment"

        recommendation.append(result)
    return recommendation

```
